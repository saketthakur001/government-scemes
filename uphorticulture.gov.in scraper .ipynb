```python
from bs4 import BeautifulSoup
import requests

def extract_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract the title
    title_element = soup.find("h1", {"id": "main-content"})
    title = title_element.text.strip() if title_element else None

    # Extract the text content with headings as keys
    data = {}
    content_div = soup.find("div", {"id": "content"})
    if content_div:
        headings = content_div.find_all(["h3"])
        for heading in headings:
            next_sibling = heading.find_next_sibling()
            while next_sibling and next_sibling.name != "h3":
                if next_sibling.string:
                    key = heading.text.strip()
                    if key not in data:
                        data[key] = ""
                    data[key] += next_sibling.string.strip() + " "
                next_sibling = next_sibling.find_next_sibling()

    return title, data

# Usage
url = ["https://uphorticulture.gov.in/en/page/rashtriya-krishi-vikas-yojna", "https://uphorticulture.gov.in/en/article/micro-irrigation?cd=NgA2AA%3D%3D", "https://uphorticulture.gov.in/en/page/scheme-is-to-promote-quality-betel-production-in-state"]

for i in url:
    title, data = extract_data(i)
        
    # Print the title
    print("Title:", title)
    print()

    # Print the extracted data
    for key, value in data.items():
        print(f"{key}:\n{value.strip()}\n")
```
